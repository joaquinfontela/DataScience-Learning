```python
import matplotlib.pyplot as plt
```

### Tenemos el siguiente sistema de ecuaciones:

#### y = 3x + 5
#### y = -x + 3
#### y = 2x + 1

```python
x = np.arange(-6,6)

y1 = 3*x + 5
y2 = -x + 3
y3 = 2*x + 1
```

```python
plt.figure()

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

plt.show()
```

#### No hay ningun punto que intercepte todas las rectas: no tiene soluciÃ³n.

### Ahora haremos un sistema solo con las dos Ãºltimas ecuaciones:

```python
plt.figure()

plt.plot(x,y2)
plt.plot(x,y3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

plt.show()
```

### Y ahora mostraremos el caso de infinitas soluciones:

```python
plt.figure()

plt.plot(x,y3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

plt.show()
```

### Sin soluciÃ³n --> Incompatible
### Una soluciÃ³n --> Compatible determinado
### Infinitas soluciones --> Compatible indeterminado
