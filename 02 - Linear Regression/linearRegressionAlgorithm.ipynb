```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_b0_and_b1(X, Y):
    n = np.size(X)
    x_mean, y_mean = np.mean(X), np.mean(Y)
    # calcule la media
    xy_sum = np.sum((X - x_mean) * (Y - y_mean))
    # obtengo la sumatoria de X-mean e Y-mean
    x_sum_squared = np.sum((X - x_mean) ** 2)
    # obtengo la sumatoria de X-mean al cuadrado.
    b1 = xy_sum/x_sum_squared
    b0 = y_mean - b1 * x_mean
    return (b0, b1)
```

```python
def plot_regression(X, Y, b):
    plt.scatter(X, Y, color = "black", marker = "o", s = 30)

    y_prediction = b[0] + b[1] * X
    plt.plot(X, y_prediction, color= "blue")

    plt.xlabel("X (independent variable)")
    plt.ylabel("Y (dependent variable)")

    plt.show()
```

```python
def main():
    X = np.array([1,2,3,4,5])
    Y = np.array([2,3,5,6,5])
    # usamos los valores del curso.
    b0, b1 = get_b0_and_b1(X, Y)
    print(b0, b1)
    # obtenemos los valores de b0 y b1 que reducen el ECM.
    plot_regression(X, Y, (b0, b1))
    # graficamos

if __name__ == "__main__":
    main()
```
