```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('caso-practico-14.csv')
```

```python
df
```

```python
df.columns = ['invoice number', 'branch', 'city', 'user type', 'gender', 'category',\
              'cost of good sold', 'sale', 'date', 'time', 'payment type', 'review']
```

```python
df
```

```python
df['weekday'] = pd.to_datetime(df['date']).dt.weekday
df
```

```python
sales_sum_for_gender_and_weekday = df.groupby(['gender', 'weekday'])['sale'].sum()
sales_sum_for_gender_and_weekday = sales_sum_for_gender_and_weekday\
                                        .to_frame()\
                                        .pivot_table(values = 'sale', index = 'gender', columns = 'weekday')
sales_sum_for_gender_and_weekday.columns = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sales_sum_for_gender_and_weekday
```

```python
sales_sum_for_gender_and_weekday.index.name = 'Gender'
```

```python
fig, ax = plt.subplots(figsize=(10,7.5))
sns.heatmap(sales_sum_for_gender_and_weekday, cmap = 'Blues',\
            ax = ax)
plt.title('2019 Q1 total sales grouped by gender and weekday')
plt.xlabel('Weekday')
plt.xticks(rotation=45)
plt.ylabel('Gender')
plt.yticks(rotation=0)
plt.show()
```

```python
df['week number'] = pd.to_datetime(df['date']).dt.isocalendar().week
df
```

```python
sales_sum_for_gender_and_date = df.groupby(['gender', 'week number'])['sale'].sum()
sales_sum_for_gender_and_date = sales_sum_for_gender_and_date\
                                        .to_frame()\
                                        .pivot_table(values = 'sale', index = 'gender', columns = 'week number')\
                                        .fillna(0)
sales_sum_for_gender_and_date
```

```python
sales_sum_for_gender_and_date.index.name = 'Gender'
sales_sum_for_gender_and_date.T.plot(figsize=(20,10), cmap='cool_r',\
                                           title = 'Sales per week and gender',\
                                           xlabel = '2019 week number', ylabel = 'Sales')
```
